```python
spy_top_10 = ['AAPL', 'MSFT', 'AMZN', 'TSLA', 'GOOGL', 'GOOG', 'FB', 'NVDA', 'BRK-B', 'JPM']
sum_lens = 0
for ticker in spy_top_10:
    data = pdr.data.DataReader(ticker, 'yahoo', '2012-1-1', '2016-1-1')
    data = processDataframe(data)
    data.replace(0.0, np.nan, inplace=True)
    data.interpolate(method='linear', limit=5, limit_area='inside', inplace=True)
    data.fillna(method='ffill', inplace=True)
    data.fillna(method='bfill', inplace=True)
    data.fillna(0, inplace=True)
    sum_lens = sum_lens + len(data)
    print(len(data))
```

```python
def processDataframe(dataframe):
    """
    GOAL: Process a downloaded dataframe to homogenize the output format.

    INPUTS:     - dataframe: Pandas dataframe to be processed.

    OUTPUTS:    - dataframe: Processed Pandas dataframe.
    """

    # Remove useless columns
    dataframe['Close'] = dataframe['Adj Close']
    del dataframe['Adj Close']

    # Adapt the dataframe index and column names
    dataframe.index.names = ['Timestamp']
    dataframe = dataframe[['Open', 'High', 'Low', 'Close', 'Volume']]

    return dataframe
```

```python
df = pd.read_csv(results_dir + 'PPO-Transfer.csv')
df
```

```python
df['Action'] = 1
df['Holdings'] = 0
df['Cash'] = 100000.
df['Money'] = 100000.
df['Returns'] = 0.
```

```python
df
```

```python
for t in range(1, len(df)):
    buy_amount = df['Cash'][t-1] * 1
    num_shares = int(buy_amount / df['Open'][t])
    df['Holdings'][t] = df['Holdings'][t-1] + num_shares
    df['Cash'][t] = df['Cash'][t-1] - num_shares * df['Open'][t]
    df['Money'][t] = df['Holdings'][t] * df['Open'][t] + df['Cash'][t]
    df['Returns'][t] = (df['Money'][t] - df['Money'][t - 1]) / df['Money'][t - 1]
```
